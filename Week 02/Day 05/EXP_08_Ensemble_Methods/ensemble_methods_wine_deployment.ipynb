# EXPERIMENT 08: Ensemble Methods (Wine Dataset Deployment)

This notebook evaluates models on the Wine dataset to make a final deployment decision.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load real tabular dataset and compare models
wine_data = load_wine()
X_wine = pd.DataFrame(wine_data.data, columns=wine_data.feature_names)
y_wine = wine_data.target

X_w_train, X_w_test, y_w_train, y_w_test = train_test_split(X_wine, y_wine, test_size=0.2, random_state=42)

dt_wine = DecisionTreeClassifier(random_state=42).fit(X_w_train, y_w_train)
rf_wine = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_w_train, y_w_train)

try:
    from xgboost import XGBClassifier
    xgb_wine = XGBClassifier(random_state=42).fit(X_w_train, y_w_train)
    xgb_acc = accuracy_score(y_w_test, xgb_wine.predict(X_w_test))
except ImportError:
    xgb_acc = "N/A (XGBoost not installed)"

print(f"Wine - Decision Tree Accuracy: {accuracy_score(y_w_test, dt_wine.predict(X_w_test)):.3f}")
print(f"Wine - Random Forest Accuracy: {accuracy_score(y_w_test, rf_wine.predict(X_w_test)):.3f}")
print(f"Wine - XGBoost Accuracy: {xgb_acc}")
```

```python
# Plot feature importances and explain top 3
importances_wine = rf_wine.feature_importances_
indices_wine = np.argsort(importances_wine)[::-1]

plt.figure(figsize=(10, 6))
plt.title("Feature Importances - Wine Dataset (Random Forest)")
plt.bar(range(X_wine.shape[1]), importances_wine[indices_wine], align="center")
plt.xticks(range(X_wine.shape[1]), X_wine.columns[indices_wine], rotation=45, ha='right')
plt.tight_layout()
plt.show()

print("Top 3 Features:")
for i in range(3):
    print(f"{i+1}. {X_wine.columns[indices_wine[i]]} (Importance: {importances_wine[indices_wine[i]]:.3f})")
```

### Explanation of Top 3 Features
The top 3 features printed above contribute the most to the Random Forest's ability to classify the type of wine. Attributes like "proline", "color_intensity", or "flavanoids" often have the strongest correlation with the specific cultivator class.

### Conclusion
**Which model to deploy and why?**
I would deploy the **Random Forest** or **XGBoost** model. A single Decision Tree is prone to overfitting and can be sensitive to noise in the data. Random Forest provides a great baseline with high stability and robust performance without requiring extensive hyperparameter tuning. If maximum accuracy is required and I have the time to tune hyperparameters, XGBoost would be the final choice. For a typical tabular dataset of this size, Random Forest's high accuracy and simplicity make it an excellent choice for production.
