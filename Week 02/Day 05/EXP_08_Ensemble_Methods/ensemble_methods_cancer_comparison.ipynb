# EXPERIMENT 08: Ensemble Methods (Breast Cancer Model Comparison)

This notebook demonstrates advanced model tuning and compares Random Forest, XGBoost, and LightGBM.


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load tabular dataset
data = load_breast_cancer()
X_bc = pd.DataFrame(data.data, columns=data.feature_names)
y_bc = data.target

X_train, X_test, y_train, y_test = train_test_split(X_bc, y_bc, test_size=0.2, random_state=42)
```

```python
# Compare Decision Tree and Random Forest
dt_bc = DecisionTreeClassifier(random_state=42).fit(X_train, y_train)
rf_bc = RandomForestClassifier(n_estimators=100, random_state=42).fit(X_train, y_train)

print(f"Breast Cancer - Decision Tree Accuracy: {accuracy_score(y_test, dt_bc.predict(X_test)):.3f}")
print(f"Breast Cancer - Random Forest Accuracy: {accuracy_score(y_test, rf_bc.predict(X_test)):.3f}")
```

```python
# Plot Feature Importances from Random Forest
importances = rf_bc.feature_importances_
indices = np.argsort(importances)[::-1][:10] # Top 10 features

plt.figure(figsize=(10, 6))
plt.title("Top 10 Feature Importances - Random Forest")
plt.bar(range(10), importances[indices], align="center")
plt.xticks(range(10), X_bc.columns[indices], rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Train XGBoost and tune hyperparameters
try:
    from xgboost import XGBClassifier
    xgb_tuned = XGBClassifier(n_estimators=150, max_depth=4, learning_rate=0.05, random_state=42)
    xgb_tuned.fit(X_train, y_train)
    print(f"Tuned XGBoost Accuracy: {accuracy_score(y_test, xgb_tuned.predict(X_test)):.3f}")
except ImportError:
    pass
```

```python
# Compare Models in a Table
try:
    import lightgbm as lgb
    lgb_bc = lgb.LGBMClassifier(random_state=42)
    lgb_bc.fit(X_train, y_train)
    lgb_acc = accuracy_score(y_test, lgb_bc.predict(X_test))
    
    from IPython.display import display
    results = pd.DataFrame({
        'Model': ['Random Forest', 'XGBoost', 'LightGBM'],
        'Accuracy': [
            accuracy_score(y_test, rf_bc.predict(X_test)),
            accuracy_score(y_test, xgb_tuned.predict(X_test)),
            lgb_acc
        ]
    })
    display(results)
except ImportError:
    print("Please run `!pip install lightgbm` to run this cell.")
```

## Viva Questions & Multiple Choice Questions

### Viva Questions
1. **What is an ensemble and why does it usually beat a single model?**
   An ensemble combines many weak models to build one strong model. It usually beats a single model because the combined vote is more robust, less prone to overfitting, and stable.
2. **What is the difference between bagging and boosting?**
   Bagging trains models in parallel on random subsets of data and averages their votes (e.g., Random Forest). Boosting trains models sequentially, where each new model fixes the previous one's mistakes (e.g., XGBoost).
3. **How does a Random Forest make a prediction?**
   It passes the input through all its decision trees. Each tree outputs a class prediction, and the Random Forest takes a majority vote to decide the final class.
4. **What does feature importance tell you?**
   It tells you which features (columns) were the most useful or contributed the most to the model's decision-making process.
5. **Why is XGBoost so popular in competitions?**
   Because it is a highly optimized, fast, and very accurate implementation of gradient boosting that often yields state-of-the-art results on tabular datasets.
6. **When would you prefer LightGBM over XGBoost?**
   LightGBM is preferred when dealing with extremely large datasets because it is highly efficient and faster than traditional XGBoost, while maintaining similar accuracy.
