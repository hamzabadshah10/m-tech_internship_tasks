# EXPERIMENT 08: Ensemble Methods (Iris Dataset Basics)

This notebook covers the basic comparison of Decision Trees and Random Forests, and tuning XGBoost.


```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# Load data
X, y = load_iris(return_X_y=True)
Xtr, Xte, ytr, yte = train_test_split(X, y, test_size=0.2, random_state=1)

# Decision Tree
dt = DecisionTreeClassifier(random_state=1).fit(Xtr, ytr)
dt_acc = accuracy_score(yte, dt.predict(Xte))

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=1).fit(Xtr, ytr)
rf_acc = accuracy_score(yte, rf.predict(Xte))

print(f"Decision Tree Accuracy: {dt_acc:.3f}")
print(f"Random Forest Accuracy: {rf_acc:.3f}")
print(f"Random Forest Feature Importances: {rf.feature_importances_.round(3)}")
```

```python
# Train XGBoost with different n_estimators and learning_rate
try:
    from xgboost import XGBClassifier
    
    # Model 1
    xgb1 = XGBClassifier(n_estimators=50, learning_rate=0.1, random_state=1)
    xgb1.fit(Xtr, ytr)
    print(f"XGBoost (n=50, lr=0.1) Accuracy: {accuracy_score(yte, xgb1.predict(Xte)):.3f}")
    
    # Model 2
    xgb2 = XGBClassifier(n_estimators=200, learning_rate=0.01, random_state=1)
    xgb2.fit(Xtr, ytr)
    print(f"XGBoost (n=200, lr=0.01) Accuracy: {accuracy_score(yte, xgb2.predict(Xte)):.3f}")
except ImportError:
    print("Please run `!pip install xgboost` to run this cell.")
```

**Observation:**
For the Iris dataset, most models will achieve near-perfect accuracy because it is a small and simple dataset. However, on more complex data, Random Forest and XGBoost generally outperform single Decision Trees because ensembles reduce variance (overfitting) and bias (in boosting).
