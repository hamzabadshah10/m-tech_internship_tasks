```
from sklearn.linear_model import LinearRegression
import numpy as np
X = np.array([[1], [2], [3], [4], [5]])
y = np.array([35, 50, 65, 80, 95])
model = LinearRegression().fit(X, y)
print(model.predict([[7], [8]]))
```
