```python
import cv2
import numpy as np
# Create a dummy image to test OpenCV
img = np.zeros((200, 200, 3), dtype=np.uint8)
cv2.rectangle(img, (50,50), (150,150), (255,0,0), -1)
cv2.imwrite('photo.jpg', img)

img_read = cv2.imread('photo.jpg')
print('Shape:', img_read.shape)
gray = cv2.cvtColor(img_read, cv2.COLOR_BGR2GRAY)
resized = cv2.resize(img_read, (128, 128))
edges = cv2.Canny(gray, 100, 200)
cv2.imwrite('edges.jpg', edges)
print('Image processing successful.')
```

```python
from tensorflow.keras import layers, models
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax'),
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
print('Total parameters:', model.count_params())
```
