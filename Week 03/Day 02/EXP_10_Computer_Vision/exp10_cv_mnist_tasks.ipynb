# Experiment 10: Computer Vision (Lab Tasks)
### Lab Tasks, Viva, MCQs


```python
import tensorflow as tf
from tensorflow.keras import layers, models
(x_train, y_train), (x_test, y_test) = tf.keras.datasets.mnist.load_data()
x_train = x_train.reshape(-1, 28, 28, 1) / 255.0
x_test = x_test.reshape(-1, 28, 28, 1) / 255.0

model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=2, verbose=1)
print('MNIST CNN Acc:', model.evaluate(x_test, y_test, verbose=0)[1])
```

### Viva Questions
1. **Color image stored?** As a 3D array (RGB).
2. **What is a pixel?** Brightness value from 0-255.
3. **Conv Layer?** Slides filters to detect patterns.
4. **Pooling?** Shrinks images to reduce compute.
5. **Classification vs Detection?** Classification labels the whole image; detection finds multiple objects.
6. **Why CNNs?** They preserve spatial relationships better than plain dense networks.
